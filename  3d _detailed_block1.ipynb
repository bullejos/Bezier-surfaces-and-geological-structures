# 3D Detailed block 1. Castillo de Mula Height.

Our next goal is to provide detailed 3D blocks of any part of the general model
representing significant regions within the modeled area, in particular we construct three
blocks. 

The topography of the general model was built using convex hulls but, in this
case, we have changed the way of doing this in each of the building blocks. Instead of
using convex hulls, we define a grid of points covering the region in a given block, and
then the grid points are colored according to the geological material they represent.
Once the topography of the block is obtained, we proceed to cut out each of the surfaces
represented in the general model to keep what is included in the block.
The most complex aspect of this process will be selecting, from the set of
points that define a surface, those that are within a given area, defined by a polygon. To
do this, the Python package Shapely will be used. This package allows
us to manipulate and analyze planar geometric objects and it is based on the widely used
open-source geometry library GEOS. Custom functions that select
points within a polygon or select the part of a BÃ©zier surface that is enclosed by a
polygon will be defined. The final results are the
Box1_3D_model.HTML, Box2_3D_model.HTML, and Box3_3D_model.HTML files.
